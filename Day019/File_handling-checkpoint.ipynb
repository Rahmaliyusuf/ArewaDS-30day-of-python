1. Write a function which count number of lines and number of words in a text. All the files are in the data the folder:
   a)  Read obama_speech.txt file and count number of lines and words
   b) Read michelle_obama_speech.txt file and count number of lines and words
   c) Read donald_speech.txt file and count number of lines and words
   d) Read melina_trump_speech.txt file and count number of lines and words

````python

def read_count_lines(x):
    with open (x) as d:
        lines_list = d.readlines()
        no_of_lines = len(lines_list)

        no_words = 0
        for i in lines_list :
            doc_words = re.sub(r'[^\w\s]','', i)
            words = doc_words.split()
            no_words += len(words)
            

        print('The number of lines in the document is ', no_of_lines)
        print('The number of words in the document is ', no_words)

print('Obama Speech')
read_count_lines('./data/obama_speech.txt')
print('Michelle Obama Speech')
read_count_lines('./data/michelle_obama_speech.txt')
print('Trump Speech')
read_count_lines('./data/donald_speech.txt')
print('Melina Trump Speech')
read_count_lines('./data/melina_trump_speech.txt')


````

#### Question 
2. Read the countries_data.json data file in data directory, create a function that finds the ten most spoken languages

````python
import json
def most_spoken_languages(filename,countries):
    file = open(filename , encoding="UTF8")
    Lang_dict =json.loads(file.read())
    languages = []
    for i in range(len(Lang_dict)):
        languages.extend(Lang_dict[i]['languages'])
    lang = {}
    for language in languages:
        lang[language] = lang.get(language,0) + 1
    sorted_lang = sorted(lang.items(), key= lambda x:x[1],reverse=True)
    top_langs = []
    for i in range(countries):
      top_langs.append(sorted_lang[i])
    return top_langs

print(most_spoken_languages('./data/countries_data.json', 10))
print(most_spoken_languages('./data/countries_data.json', 3))
````

````python
def most_populated_countries(filename,countries):
    file = open(filename , encoding="UTF8")
    Lang_dict =json.loads(file.read())
    population = {}
    for i in range(len(Lang_dict)):
        keys = Lang_dict[i]['name']
        values = Lang_dict[i]['population']
        population[keys] = values
    sorted_pop = dict(sorted(population.items(), key= lambda x:x[1],reverse=True))

    most_pop = []
    for i in list(sorted_pop.items()) [:countries] :
        most_pop.append({"Country" : i[0], "Population" : i[1]})
    return most_pop

print(most_populated_countries('./data/countries_data.json', 10))
most_populated_countries('./data/countries_data.json', 3)
````

##### Question 5
Find the most common words in the English language. Call the name of your function find_most_common_words, it will take two parameters - a string or a file and a positive integer, indicating the number of words. Your function will return an array of tuples in descending order. Check the output
```py
    # Your output should look like this

    (6, 'to'),

    (2, 'have'),
    (2, 'I')]

    # Your output should look like this
    print(find_most_common_words('sample.txt', 5))

    [(10, 'the'),
    (8, 'be'),
    (6, 'to'),
    (6, 'of'),
    (5, 'and')]
```

````python
def find_most_common_words(filename,top_words):
    with open(filename) as file:
        lines = file.readlines()
    words = []
    for line in lines:
        line = re.sub(r'[^\w\s]','',line)
        line_no_num = re.sub(r'\d','',line)
        words.extend(line_no_num.split())
    words_dict = {}
    for word in words:
        words_dict[word] = words_dict.get(word,0) + 1
    words_sorted = sorted(words_dict.items(),key=lambda x:x[1],reverse=True)
    result = [(word[1],word[0]) for word in words_sorted]
    return result[:top_words]

find_most_common_words('./sample.txt', 10)
````

##### Question
6. Use the function, find_most_frequent_words to find:
   a) The ten most frequent words used in Obama's speech
   b) The ten most frequent words used in Michelle's speech
   c) The ten most frequent words used in Trump's speech
   d) The ten most frequent words used in Melina's speech

````python
def find_most_common_words(filename,top_words=10):
    with open(filename) as file:
        lines = file.readlines()
    words = []
    for line in lines:
        line = re.sub(r'[^\w\s]','',line)
        line_no_num = re.sub(r'\d','',line)
        words.extend(line_no_num.split())
    words_dict = {}
    for word in words:
        words_dict[word] = words_dict.get(word,0) + 1
    words_sorted = sorted(words_dict.items(),key=lambda x:x[1],reverse=True)
    result = [(word[1],word[0]) for word in words_sorted]
    return result[:top_words]

print('Ten most common words in Obama speech are : ',find_most_common_words('./data/obama_speech.txt'))
print('Ten most common words in Michelle Obama speech are : ',find_most_common_words('./data/michelle_obama_speech.txt'))
print('Ten most common words in Donald Trump speech are : ',find_most_common_words('./data/donald_speech.txt'))
print('Ten most common words in Melina Trump speech are : ',find_most_common_words('./data/melina_trump_speech.txt'))
````
