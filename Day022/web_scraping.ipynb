##### Question 1 Scrape the following website and store the data as json file(url = 'http://www.bu.edu/president/boston-university-facts-stats/').

```python
import requests
from bs4 import BeautifulSoup
import json

url = 'http://www.bu.edu/president/boston-university-facts-stats/'
response = requests.get(url)
content = response.content
soup = BeautifulSoup(content, 'html.parser')
Boston = soup.body.get_text()
print(soup.body.get_text())
with open('boston_facts.json','w') as f:
    json.dump(Boston,f)
    
```

##### Question 2 
 Extract the table in this url (https://archive.ics.uci.edu/ml/datasets.php) and change it to a json file

```python
url = 'https://archive.ics.uci.edu/ml/datasets.php'

response = requests.get(url)
content = response.content 
soup = BeautifulSoup(content, 'html.parser')  


tables = soup.find_all('table', {'cellpadding':'3'})
table = tables[0]
table_json = json.dumps(table)
with open('dataset_table.json','w') as f:
    json.dump(table_json, f)
    
```
